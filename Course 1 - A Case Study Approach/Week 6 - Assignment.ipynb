```python
import graphlab as gl
```

```python
train_data = gl.SFrame('image_train_data/')
```

```python
train_data['label'].sketch_summary()
```

```python
cars = train_data[train_data['label'] == 'automobile']
cats = train_data[train_data['label'] == 'cat']
dogs = train_data[train_data['label'] == 'dog']
birds = train_data[train_data['label'] == 'bird']
```

```python
cars_model = gl.nearest_neighbors.create(cars, features = ['deep_features'], label = 'id')
cats_model = gl.nearest_neighbors.create(cats, features = ['deep_features'], label = 'id')
dogs_model = gl.nearest_neighbors.create(dogs, features = ['deep_features'], label = 'id')
birds_model = gl.nearest_neighbors.create(birds, features = ['deep_features'], label = 'id')
```

```python
test_data = gl.SFrame('image_test_data/')
```

```python
gl.canvas.set_target('ipynb')
```

```python
cats_model.query(test_data[0:1])
```

```python
def get_images_from_ids(query_result):
    return train_data.filter_by(query_result['reference_label'],'id')
```

```python
cats_images = get_images_from_ids(cats_model.query(test_data[0:1]))
cats_images.show()
```

```python
dogs_images = get_images_from_ids(dogs_model.query(test_data[0:1]))
dogs_images.show()
```

```python
image_test_cat = test_data[test_data['label'] == 'cat']
image_test_dog = test_data[test_data['label'] == 'dog']
image_test_bird = test_data[test_data['label'] == 'bird']
image_test_automobile = test_data[test_data['label'] == 'automobile']
```

```python
dog_dog_neighbors = dogs_model.query(image_test_dog, k = 1)
dog_cat_neighbors = cats_model.query(image_test_dog, k = 1)
dog_car_neighbors = cars_model.query(image_test_dog, k = 1)
dog_bird_neighbors = birds_model.query(image_test_dog, k = 1)

cat_cat_neighbors = cats_model.query(image_test_cat, k = 1)
cat_dog_neighbors = dogs_model.query(image_test_cat, k = 1)
cat_bird_neighbors = birds_model.query(image_test_cat, k = 1)
cat_car_neighbors = cars_model.query(image_test_cat, k = 1)

car_car_neighbors = cars_model.query(image_test_automobile, k = 1)
car_dog_neighbors = dogs_model.query(image_test_automobile, k = 1)
car_bird_neighbors = birds_model.query(image_test_automobile, k = 1)
car_cat_neighbors = cats_model.query(image_test_automobile, k = 1)

bird_bird_neighbors = birds_model.query(image_test_bird, k = 1)
bird_dog_neighbors = dogs_model.query(image_test_bird, k = 1)
bird_cat_neighbors = cats_model.query(image_test_bird, k = 1)
bird_car_neighbors = cars_model.query(image_test_bird, k = 1)
```

```python
dis_table = gl.SFrame()
```

```python
dis_table['dog-automobile'] = dog_car_neighbors['distance']
dis_table['dog-bird'] = dog_bird_neighbors['distance']
dis_table['dog-cat'] = dog_cat_neighbors['distance']
dis_table['dog-dog'] = dog_dog_neighbors['distance']
```

```python
dis_table.head()
```

```python
def is_dog_correct(tab):
    if tab['dog-dog'] < tab['dog-cat'] and tab['dog-dog'] < tab['dog-automobile'] and tab['dog-dog'] < tab['dog-bird']:
        return 1
    return 0
```

```python
table = dis_table.apply(is_dog_correct)
```
